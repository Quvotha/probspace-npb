# 選手にIDを割り振る

```python
from inning import Inning
from teams import Teams
from players import Players
```

```python
train[['inningNo', 'isBottom']] = train.inning.apply(Inning.extract_info)
train[['inning', 'inningNo', 'isBottom']]
```

```python
test[['inningNo', 'isBottom']] = test.inning.apply(Inning.extract_info)
test[['inning', 'inningNo', 'isBottom']]
```

```python
train['batterTeam'] = train.apply(Teams.extract_batter_team, axis=1)
test['batterTeam'] = test.apply(Teams.extract_batter_team, axis=1)
```

```python
train['pitcherTeam'] = train.apply(Teams.extract_pitcher_team, axis=1)
test['pitcherTeam'] = test.apply(Teams.extract_pitcher_team, axis=1)
```

```python
Teams.get_team_map()
```

```python
pitchers = pd.DataFrame({
    'pitcher'    : pd.concat([train.pitcher, test.pitcher], axis=0),
    'pitcherTeam': pd.concat([train.pitcherTeam, test.pitcherTeam], axis=0),
})
pitchers
```

```python
batters = pd.DataFrame({
    'batter'    : pd.concat([train.batter, test.batter]),
    'batterTeam': pd.concat([train.batterTeam, test.batterTeam]),
})
batters
```

```python
%%time
players = Players(pitchers=pitchers, batters=batters)
player_id = players.assign_id()
player_id
```

```python
for p in players.TRANSFERRED_PLAYERS:
    df_ = player_id.query(f'player == "{p}"')
    display(df_)
del df_
```

```python
nrow_train, nrow_test = train.shape[0], test.shape[0]
train = pd.merge(train,
                 player_id.rename(columns={'playerID': 'batterID'}),
                 left_on=['batter', 'batterTeam'],
                 right_on=['player', 'team'],
                 how='inner')
train = pd.merge(train,
                 player_id.rename(columns={'playerID': 'pitcherID'}),
                 left_on=['pitcher', 'pitcherTeam'],
                 right_on=['player', 'team'],
                 how='inner')
train.drop(columns=['team_x', 'player_x', 'team_y', 'player_y'], inplace=True)
assert(train.shape[0] == nrow_train)


test = pd.merge(test,
                player_id.rename(columns={'playerID': 'batterID'}),
                left_on=['batter', 'batterTeam'],
                right_on=['player', 'team'],
                how='left')
test = pd.merge(test,
                player_id.rename(columns={'playerID': 'pitcherID'}),
                left_on=['pitcher', 'pitcherTeam'],
                right_on=['player', 'team'],
                how='left')
test.drop(columns=['team_x', 'player_x', 'team_y', 'player_y'], inplace=True)
assert(test.shape[0] == nrow_test)
```
